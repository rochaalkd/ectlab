```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
df=pd.DataFrame({
    'Frequency':[1,3,5,7,10,30,50,70,100,500,700,1000],
    'Output Amplitude':[1.84,1.82,1.77,1.72,1.62,0.96,0.616,0.44,0.292,0.0464,0.0624,0.08],
    'Time':[10,9.6,9,8.8,8.6,5.8,4.16,3.2,2.42,0.12,0.1,0.07],
    'Input Amplitude':[1.91,1.94,1.91,1.96,1.96,2,2,2,2,1.96,1.96,1.96]
})
```

```python
df
```

```python
df['Frequency']=1000*df['Frequency']
```

```python
df['Gain']=20*(np.log10(df['Output Amplitude']/df['Input Amplitude']))
```

```python
df['Gain']
```

```python
df['Time(ms)']=df['Time']*1*math.pow(10,-6)
```

```python
df['Time(ms)']
```

```python
df['Angle(degree)']=360*df['Frequency']*df['Time(ms)']
```

```python
df['Angle(degree)']
```

```python
plt.semilogx(df['Frequency'],df['Gain'])
plt.grid()
```

```python
df3=pd.DataFrame({
    'Frequency':[1,3,7,10,30,50,70,100,500,1000],
    'OutputVolt':[0.126,0.375,0.81,1.075,1.75,1.9,1.95,1.975,2.00,2.00],
    'Time':[240,72.5,26.25,16.0,2.38,0.9,0.52,0.27,0,0],
    'InputVolt':[2,2,2,2,2,2,2,2,2,2]
})
```

```python
df3['Gain']=20*np.log10(df3['OutputVolt']/2)
```

```python
df3['Gain']
```

```python
df3['FREQ']=1000*df3['Frequency']
```

```python
df3
```

```python
plt.figure(figsize=(15,5))
plt.semilogx(df3['FREQ'],df3['Gain'])
plt.hlines(-3,0,1000000,linestyles='dotted')
plt.vlines(18700,0,-25,linestyles='dotted')
plt.legend(['curve','vline=18700','hline=-3'],loc='lower right')
plt.xlabel('Frequency(Hz)')
plt.ylabel('Transferfunction(dB)')
plt.grid()
plt.savefig('Highpassfreq.jpeg')
```
